```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch.nn as nn
import torch
```

```python
def relu(x):
    f = nn.ReLU()
    return f(torch.tensor(x, dtype=torch.float32)).numpy()

def gelu(x):
    f = nn.GELU()
    return f(torch.tensor(x, dtype=torch.float32)).numpy()

def hardswish(x):
    f = nn.Hardswish()
    return f(torch.tensor(x, dtype=torch.float32)).numpy()

def hardsigmoid(x):
    f = nn.Hardsigmoid()
    return f(torch.tensor(x, dtype=torch.float32)).numpy()

def swish(x):
    f = nn.SiLU()
    return f(torch.tensor(x, dtype=torch.float32)).numpy()

def sigmoid(x):
    f = nn.Sigmoid()
    return f(torch.tensor(x, dtype=torch.float32)).numpy()
```

```python
x = np.arange(start=-4, stop=4, step=0.1)
y_r = relu(x)
y_g = gelu(x)
y_hs = hardswish(x)
y_s = swish(x)
y_sig = sigmoid(x)
y_hsig = hardsigmoid(x)
```

```python
sns.set(font_scale=1.25)
sns.set_style("whitegrid", {'axes.grid' : True})
```

```python
ax = sns.lineplot(x=x, y=y_r)
ax.set(xlabel="x", ylabel="ReLU(x)")
plt.title("ReLU(x) = max(0, x)")
plt.savefig("relu.eps")
```

```python
temp = pd.DataFrame({"x": x, "ReLU": y_r, "GELU": y_g})
temp = pd.melt(temp, ["x"])
temp = temp.rename(columns={"variable": "função"})
ax = sns.lineplot(data=temp, x="x", y="value", hue="função")
ax.set_title("ReLU vs. GELU")
ax.set(xlabel="x")
ax.set(ylabel="f(x)")
plt.savefig("reluvsgelu.eps")
```

```python
temp = pd.DataFrame({"x": x, "Hardsigmoid": y_hsig, "Sigmoid": y_sig})
temp = pd.melt(temp, ["x"])
temp = temp.rename(columns={"variable": "função"})
ax = sns.lineplot(data=temp, x="x", y="value", hue="função")
ax.set_title("Hardsigmoid vs. Sigmoid")
ax.set(xlabel="x")
ax.set(ylabel="f(x)")
plt.savefig("hardsigmoidvssigmoid.eps")
```

```python
temp = pd.DataFrame({"x": x, "Hardswish": y_hs, "Swish": y_s})
temp = pd.melt(temp, ["x"])
temp = temp.rename(columns={"variable": "função"})
ax = sns.lineplot(data=temp, x="x", y="value", hue="função")
ax.set_title("Hardswish vs. Swish")
ax.set(xlabel="x")
ax.set(ylabel="f(x)")
plt.savefig("hardswishvsswish.eps")
```
